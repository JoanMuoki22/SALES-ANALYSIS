```python
#Importation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
#plt.style.use('fivethirtyeight')
#ptl.style.use('')
df=pd.read_csv('Sample - Superstore.csv')

df.head(2)
```

```python
#Explore Data Distribution
#Produce summary statistics
#df.info()
df.describe()
# df.groupby('State').Profit.agg(['sum','mean','min','max','count','median','std','var'])
# df.groupby('City').Profit.agg(['sum','mean','min','max','count','median','std','var'])
# df.isnull().sum()
```

```python
#Ship Mode Distribution
ship_mode_counts = df['Ship Mode'].value_counts()
ship_mode_counts.plot(kind='bar')
plt.title('Ship Mode Distribution')
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.show()
```

```python
#Segment Distribution
segment_counts = df['Segment'].value_counts()
segment_counts.plot(kind='bar')
plt.title('Segment Distribution')
plt.xlabel('Segment')
plt.ylabel('Count')
plt.show()
```

```python
#City Distribution (Top 10 cities):
top_cities = df['City'].value_counts().head(10)
top_cities.plot(kind='bar')
plt.title('Top 10 Cities Distribution')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Sales Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Sales'], bins=50, kde=True)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
#Quantity Distribution:
plt.figure(figsize=(8, 6))
sns.histplot(df['Quantity'], bins=20, kde=True)
plt.title('Quantity Distribution')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
#Profit Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Profit'], bins=50, kde=True)
plt.title('Profit Distribution')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()
```
