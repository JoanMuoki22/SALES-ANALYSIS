```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load the dataset
df = pd.read_csv('Sample - Superstore.csv')
df
```

```python

# Task 1: Explore Data Distribution
# Summary statistics for relevant columns
summary_stats = df[['Ship Mode', 'Segment', 'City', 'Sales', 'Quantity', 'Discount', 'Profit']].describe()

# Visualizations
plt.figure(figsize=(15, 8))

# Visualize Ship Mode distribution
plt.subplot(231)
sns.countplot(data=df, x='Ship Mode')

# Visualize Segment distribution
plt.subplot(232)
sns.countplot(data=df, x='Segment')

# Visualize City distribution (consider top cities)
plt.subplot(233)
top_cities = df['City'].value_counts().head(10)
sns.barplot(x=top_cities.index, y=top_cities.values)
plt.xticks(rotation=90)

# Visualize Sales, Quantity, Discount, and Profit distributions
plt.subplot(234)
sns.histplot(data=df, x='Sales', bins=30, kde=True)
plt.subplot(235)
sns.histplot(data=df, x='Quantity', bins=30, kde=True)
plt.subplot(236)
sns.histplot(data=df, x='Discount', bins=30, kde=True)

plt.show()

# Interpret the distributions in text (based on the summary statistics and visualizations)



# Task 3: Explore correlations
correlation_matrix = df.corr()

# Find features with the strongest positive and negative correlations with Sales
strong_positive_corr = correlation_matrix['Sales'].sort_values(ascending=False).head(3)
strong_negative_corr = correlation_matrix['Sales'].sort_values(ascending=True).head(3)

# Visualize the relationships
sns.heatmap(data=correlation_matrix, annot=True, cmap='coolwarm')



```

```python
# Task 2: Explore Differences between Subsets
# Create subsets based on Quantity and City
subset1 = df[df['Quantity'] < 5]
subset2 = df[df['Quantity'] >= 5]

# Visualize Sales distribution for each subset
plt.figure(figsize=(12, 5))
plt.subplot(121)
sns.histplot(data=subset1, x='Sales', bins=30, kde=True, color='blue', label='<5 Quantity')
plt.subplot(122)
sns.histplot(data=subset2, x='Sales', bins=30, kde=True, color='green', label='>=5 Quantity')
plt.legend()
plt.show()
```

```python
# Task 4: Engineer and Explore a New Feature
# Create a new feature as the difference between Profit and Discount
df['Profit_Discount_Diff'] = df['Profit'] - df['Discount']

# Visualize the relationship between the new feature and Sales
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Profit_Discount_Diff', y='Sales', alpha=0.5)
```
